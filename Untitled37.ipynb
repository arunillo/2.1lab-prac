```python
%%writefile text_analyzer.py
import string


def analyze_text(raw_text: str) -> dict[str, int | float]:
    """
    Анализирует текст и возвращает статистику.

    Аргументы:
        raw_text: исходный текст (может содержать знаки препинания,
                  пробелы, разный регистр)

    Возвращает:
        Словарь с ключами:
            'total_words' (int) — общее количество слов,
            'unique_words' (int) — количество уникальных слов,
            'avg_word_length' (float) — средняя длина слова
                                        (округлено до 2 знаков)
    """
    # Приводим к нижнему регистру и удаляем все знаки препинания
    translator = str.maketrans('', '', string.punctuation)
    cleaned = (
        raw_text.lower()
        .translate(translator)
    )

    # Разбиваем на слова (пустые строки автоматически отбрасываются)
    words = cleaned.split()

    total = len(words)
    if total == 0:
        return {'total_words': 0, 'unique_words': 0, 'avg_word_length': 0.0}

    unique = len(set(words))
    avg = round(sum(len(w) for w in words) / total, 2)

    return {
        'total_words': total,
        'unique_words': unique,
        'avg_word_length': avg
    }


def main():
    """Демонстрация работы функции analyze_text."""
    raw_text = "Hello, world! Hello everyone. This is a test... A TEST!"
    stats = analyze_text(raw_text)
    print(stats)


if __name__ == "__main__":
    main()
```
