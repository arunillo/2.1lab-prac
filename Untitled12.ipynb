```python

#Лабораторная работа №1: Культура кода и Типизация

import string
from typing import Dict, Union

 #Анализирует текст и возвращает статистику.
def analyze_text(raw_text: str) -> Dict[str, Union[int, float]]:

    # Приведение к нижнему регистру
    text_lower = raw_text.lower()

    # Удаление знаков препинания
    # Используем translate для эффективности
    translator = str.maketrans('', '', string.punctuation)
    text_clean = text_lower.translate(translator)

    # Разделение на слова
    words = text_clean.split()

    # Вычисление статистики
    total_words = len(words)
    unique_words = len(set(words))

    if total_words > 0:
        avg_word_length = round(sum(len(word) for word in words) / total_words, 2)
    else:
        avg_word_length = 0.0

    return {
        'total_words': total_words,
        'unique_words': unique_words,
        'avg_word_length': avg_word_length
    }


# Пример использования (можно удалить или оставить для самопроверки)
if __name__ == "__main__":
    test_texts = [
        "Hello, world! Hello everyone.",
        "   Python   is  awesome!!!   And   it's   great   ",
        ""
    ]

    for text in test_texts:
        print(f"Текст: {text!r}")
        print(f"Статистика: {analyze_text(text)}\n")
```
